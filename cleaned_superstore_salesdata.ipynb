# Standardize Column Names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df
```

# Check and Handle Missing Values

```python
df.isnull().sum()
```

```python
df['postal_code'] = df['postal_code'].fillna(0).astype(int)
df
```

#  Checking Duplicates 

```python
df.duplicated().sum()
```

# Convert Data Types

```python
df['order_date'] = pd.to_datetime(df['order_date'],format='%d/%m/%Y')
df['ship_date'] = pd.to_datetime(df['ship_date'],format = '%d/%m/%Y')
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df
```

# Create New Columns

```python
df['shipping_duration'] = (df['ship_date'] - df['order_date']).dt.days
df
```

```python
df['order_month'] = df['order_date'].dt.to_period('M')
df
```

# Standardize Text Columns

```python
text_columns = ['ship_mode', 'customer_name', 'segment', 'country', 'city',
                'state', 'region', 'category', 'sub-category', 'product_name']

for col in text_columns:
    df[col] = df[col].str.strip().str.lower()
df
```

# Drop Unnecessary Columns

```python
df.drop(columns=['row_id'])
```

```python
df.to_csv('cleaned_superstore_sales', index=False)
```
